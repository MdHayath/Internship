```python
import requests
from bs4 import BeautifulSoup
import re  
import pandas as pd
import numpy as np
from lxml import etree
import time

Company = []
Reviews = []
Rating = []
Verified= []
Web_site = []
Contact_no = []
Address = []
About = []

count = 0
for i in range(1,2):
    url = "https://www.goodfirms.co/directory/platform/app-development?page="+str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text,"html.parser")
    links = soup.find_all("a",class_="profile-link")

#     REVIEWS
    reviews = soup.find_all('span',class_="review-count")
    for i in reviews:
        if i is not None:
            a = i.text.strip()
            Reviews.append(a)
        else:
            Reviews.append(np.nan)
#     RATING
    rating = soup.find_all("span", class_="review-rating")
    for i in rating:
        if i is not None:
            a = i.text.strip()
            new = a.split()
            Rating.append(new[0])
        else:
            Rating.append(np.nan)
#    NAME OF THE COMPANY
    links = soup.find_all("a",class_="profile-link")
    for i in links:
        name = str(i.get("href"))
        if name is not None:
            name1 = name.replace('https://www.goodfirms.co/company/','')
            Company.append(name1)
        else:
             Company.append(np.nan)
#     WEBSITE
        web = str(i.get("href"))
        if web is not None:
            Web_site.append(web)
        else:
            Web_site.append(np.nan)

                
    
    
"""#         VERIFIED_STATUS
    Verify = str(soup.find_all('span',class_="custom_tooltip"))
    Status = re.findall(r'Verified',Verify)
    for i in Status:
        if i is not None:
            Verified.append(i)
        else:
            Verified.append(np.nan)"""    
"""      
```
